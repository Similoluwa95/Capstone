```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Setting up Variables for Linear Regression Test 1
df2 = pd.read_csv("Ecommerce_Sales.csv")
```

```python
df2["Product_Category"].value_counts()
```

```python
df2["Product_Category"].replace("Electronics",1,inplace=True)
df2["Product_Category"].replace("Sports",2,inplace=True)
df2["Product_Category"].replace("Toys",3,inplace=True)
df2["Product_Category"].replace("Home Decor",4,inplace=True)
df2["Product_Category"].replace("Fashion",5,inplace=True)
```

```python
X = df2[["Product_Category","Price","Discount","Marketing_Spend"]]
y = df2["Profit"]
```

```python
# Multivariate Regression (Train/Test Split)

X_train,X_test,y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=1)
```

```python
# Instantiate a LinearRegression Model
model = LinearRegression()

# Fit training data to model
model.fit(X_train,y_train)
```

```python
# Generate y_pred
y_pred = model.predict(X_test)

# Evaluate Model
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(mse)
print(r2)
```
